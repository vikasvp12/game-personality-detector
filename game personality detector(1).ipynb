```
!pip install transformers --quiet

import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

texts = [
    "Let's rush and attack now!",
    "Stay back, wait for the right time.",
    "I love playing with the team.",
    "I prefer solo missions."
]
labels = [0, 1, 2, 3]


label_names = {
    0: "Aggressive Rusher",
    1: "Defensive Planner",
    2: "Team Supporter",
    3: "Solo Strategist"
}

model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=4)

test_text = "Woo we won, I carried the match with my strategy!"
inputs = tokenizer(test_text, return_tensors="pt", truncation=True, padding=True)
outputs = model(**inputs)
predicted_id = torch.argmax(outputs.logits).item()

print("Input:", test_text)
print("Predicted Personality Type:", label_names[predicted_id])
```
